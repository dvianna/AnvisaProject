```python
import pandas as pd
from urllib.request import urlretrieve
import time
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
```

```python
df = pd.read_excel('./ExportPortalAntigo.xlsx', sheet_name='Planilha1')
```

```python
df.describe()
```

```python
link_list = df['ns1:link-resolucao'].to_list()
```

```python
count = 1
previous_link = ''
for link in link_list:
    try:
        if 'http://pesquisa.in.gov.br/imprensa/jsp/' in link:
            continue
        if link != previous_link:
            print(count)
            previous_link = link
            path = './data/' + str(count)
            urlretrieve(link, path)
            time.sleep(10)
            count += 1
    except:
        print(link)        
        continue
```
